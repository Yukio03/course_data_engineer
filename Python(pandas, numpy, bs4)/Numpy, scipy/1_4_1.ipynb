```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh
%matplotlib inline
```

```python
r = sps.rayleigh()
```

```python
def show_pdf(pdf, xmin, xmax, ymax, grid_size, distr_name, **kwargs):
  grid = np.linspace(xmin, xmax, grid_size)
  plt.figure(figsize=(12, 5))
  plt.plot(grid, pdf(grid, **kwargs), lw=5)
  plt.grid(ls=':')
  plt.xlabel('Значение', fontsize=18)
  plt.ylabel('Плотность', fontsize=18)
  plt.xlim((xmin, xmax))
  plt.ylim((None, ymax))
  title = f'Плотность {distr_name}'
  plt.title(title.format(**kwargs), fontsize=20)
  plt.show()
```

```python
show_pdf(pdf = sps.rayleigh.pdf, xmin = 0, xmax = 3, ymax = 1, grid_size=100, distr_name=r'$N({loc}, {scale})$', loc=0, scale=1)
```

```python
sample = r.rvs(size = 1000)
grid = np.linspace(0, 3, 1000)

plt.figure(figsize=(12, 5))
plt.hist(sample, bins=15, density=True, alpha=0.6)
plt.plot(grid, r.pdf(grid), color='blue', lw=5)
plt.grid(ls=':')
plt.show()
```
