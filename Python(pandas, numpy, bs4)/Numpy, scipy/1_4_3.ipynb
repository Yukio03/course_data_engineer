```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
```

```python
with open("dataset_home.txt", mode="r") as file:
    for human in file.readlines():
        age.append(int(human.split(",")[0]))

        if human.split(",")[1] == "male":
          sex.append(1)
        else:
          sex.append(0)

        bmi.append(float(human.split(",")[2]))

        children.append(int(human.split(",")[3]))

        if human.split(",")[4] == "yes":
          smoker.append (1)
        else:
          smoker.append (0)

        if human.split(",")[5] == "northwest":
          region.append(0)
        elif human.split(",")[5] == "northeast":
          region.append(1)
        elif human.split(",")[5] == "southwest":
          region.append(2)
        elif human.split(",")[5] == "southeast":
          region.append(3)
        
        charges.append(float(human.split(",")[6]))
```

```python
age = np.array(age) # возраст
sex = np.array(sex) # пол
bmi = np.array(bmi) # индекс массы тела
children = np.array(children) # кол-во детей
smoker = np.array(smoker)
region = np.array(region)
charges = np.array(charges)
```

**Корреляция**

```python
plt.scatter(np.log(charges), age)
# бюджет страховки и возраст
# Если честно, не смог определить корреляцию, но на этом графике вроде что-то похожее
```
