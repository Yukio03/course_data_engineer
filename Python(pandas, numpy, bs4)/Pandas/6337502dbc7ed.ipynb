```python
import pandas as pd
```

```python
data = pd.read_csv("ds_salaries.csv", index_col = 'Unnamed: 0')
data
```

Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)

```python
data.groupby('job_title')
```

Какая средняя и медианная зарплата по группам вакансий?

```python
print(data.groupby('job_title').salary_in_usd.mean())
print(data.groupby('job_title').salary_in_usd.median())
```

Какая средняя и медианная зарплата по каждому региону?

```python
print(data.groupby('company_location').salary_in_usd.mean())
print(data.groupby('company_location').salary_in_usd.median())
```

Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
max(data.groupby('job_title').salary_in_usd.mean())
```

Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
data['company_location'].value_counts().head()
```

```python
keys = data['company_location'].value_counts().keys() # индексы массива для перебора
arr = data['company_location'].value_counts() # сам массив 
arr
```

```python
count_company = data['company_location'].value_counts().sum()
for ind in keys:
  print(ind, round(arr[ind]*100/count_company, 2))
```
